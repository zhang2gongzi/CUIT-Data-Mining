```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载CSV文件
df = pd.read_csv('年度数据房价平均销售价格_转置.csv', encoding='gbk')

# 使用已加载的数据框
data = df

# 提取年份作为 X 轴
years = data['年份']

# 删除 '年份' 列，只保留价格数据
price_data = data.drop(columns=['年份'])

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制每种类型的折线图
for column in price_data.columns:
    plt.plot(years, price_data[column], label=column)
# 设置 X 轴刻度，显示每年的数据
plt.xticks(years, rotation=45)  # 旋转45度以防止标签重叠

# 添加标题和轴标签
plt.title('Trend of average sales prices of various types of commercial housing over the years')
plt.xlabel('year')
plt.ylabel('price/m^2')

# 添加图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图表
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 提供的数据点
x_data = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y_data = [5576, 5833.95, 6584.93, 7471.25, 7801, 9662, 10934, 10993.92, 11460.19, 12591, 12965, 15157, 15911, 14965, 16252, 17886]

# 将数据转换为NumPy数组
X_train = np.array(x_data).reshape(-1, 1)
y_train = np.array(y_data).reshape(-1, 1)

# 创建并训练线性回归模型
model = LinearRegression()
model.fit(X_train, y_train)

# 预测未来的两个数据点
future_points = np.array([2020, 2021,2022]).reshape(-1, 1)
predictions = model.predict(future_points)

print("预测的未来两个数据点:", predictions.flatten())
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 加载数据
df1 = pd.read_csv('年度数据房价平均销售价格_转置.csv', encoding='gbk')
df2 = pd.read_csv('年度数据国内生产总值_转置.csv', encoding='gbk')

# 合并数据
merged_df = pd.merge(df1, df2, on='year', how='inner')

# 处理缺失值
merged_df.dropna(inplace=True)

# 检查数据长度
print(len(merged_df['average']))

# 数据标准化
scaler = StandardScaler()
scaled_data = scaler.fit_transform(merged_df.drop(columns=['year']))
scaled_df = pd.DataFrame(scaled_data, columns=merged_df.columns[1:])
scaled_df['year'] = merged_df['year']

# 再次处理缺失值
scaled_df.dropna(inplace=True)

# 检查数据长度
print(len(scaled_df['average']))

# 计算相关矩阵
correlation_matrix = scaled_df.corr()
print(correlation_matrix)

# ACF图
sm.graphics.tsa.plot_acf(scaled_df['average'], lags=18)
plt.show()

# PACF图
sm.graphics.tsa.plot_pacf(scaled_df['average'], lags=18)
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
# 加载数据
df1 = pd.read_csv('年度数据房价平均销售价格_转置.csv', encoding='gbk')
df2 = pd.read_csv('年度数据国内生产总值_转置.csv', encoding='gbk')

# 合并数据
merged_df = pd.merge(df1, df2, on='year', how='inner')

# 处理缺失值
merged_df.dropna(inplace=True)


# 假设 df 是包含时间序列数据的 DataFrame，且 'average' 列是我们关注的目标
#df = pd.read_csv('your_data_file.csv')  # 示例数据加载，请替换为实际数据路径

# 使用 'average' 列作为时间序列数据
time_series = merged_df['average'].values

# 绘制 ACF 图
fig, ax = plt.subplots(figsize=(14, 6))
sm.graphics.tsa.plot_acf(time_series, lags=40, ax=ax)
ax.set_title('Autocorrelation Function (ACF)')
plt.show()

# 绘制 PACF 图
fig, ax = plt.subplots(figsize=(14, 6))
sm.graphics.tsa.plot_pacf(time_series, lags=40, ax=ax)
ax.set_title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 加载数据
df1 = pd.read_csv('年度数据房价平均销售价格_转置.csv', encoding='gbk')
df2 = pd.read_csv('年度数据国内生产总值_转置.csv', encoding='gbk')

# 合并数据
merged_df = pd.merge(df1, df2, on='year', how='inner')

# 处理缺失值
merged_df.dropna(inplace=True)

# 假设df1和df2已经被正确地加载和合并
# 现在我们想要计算df2的所有列与df1中'average'列的相关系数

# 首先，提取df1中的'average'列
average_series = merged_df['average']

# 然后，对于df2中的每一列，计算其与average_series的相关系数
for column in df2.columns:
    if column != 'year':  # 如果不是'year'列
        correlation = average_series.corr(df2[column])
        print(f"Correlation between {column} and average: {correlation}")

# 提取房价平均销售价格的时间序列数据
time_series = merged_df['average']

# 检查时间序列的长度
print(f"Length of time series: {len(time_series)}")

# 根据时间序列的长度动态设置lags参数
lags = min((len(time_series) // 2) - 1, 40)

# 绘制 ACF 图
fig, ax = plt.subplots(figsize=(14, 6))
sm.graphics.tsa.plot_acf(time_series, lags=lags, ax=ax)
ax.set_title('Autocorrelation Function (ACF) of Average Housing Sale Price')
plt.show()

# 绘制 PACF 图
fig, ax = plt.subplots(figsize=(14, 6))
sm.graphics.tsa.plot_pacf(time_series, lags=lags, ax=ax)
ax.set_title('Partial Autocorrelation Function (PACF) of Average Housing Sale Price')
plt.show()
```
